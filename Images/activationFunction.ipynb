```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax1.set_aspect(5.0)
plt.title('Sigmoid function')
plt.xlabel('x')
plt.ylabel('A(x)')
plt.grid(True)

x = np.arange(-5, 5, 0.05)
y1 = []
y2 = []

for xx in x:
  y1.append(1/(1+exp(-xx)))
  y2.append(max(0,xx))

plt.plot(x, y1)

ax2 = fig.add_subplot(1,2,2)
ax2.set_aspect(1.0)
plt.title('ReLU')
plt.xlabel('x')
plt.ylabel('A(x)')
plt.grid(True)
plt.plot(x, y2)

fig.savefig('activation_function')
```

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax1.set_aspect(20.0)
plt.title('Differentiation of sigmoid function.png')
plt.xlabel('x')
plt.ylabel('A(x)')
plt.grid(True)

x = np.arange(-5, 5, 0.05)
y1 = []
y2 = []

for xx in x:
  y1.append(exp(-xx)/((1+exp(-xx))*(1+exp(-xx))))
  y2.append(0 if xx<0 else 1)

plt.plot(x, y1)

ax2 = fig.add_subplot(1,2,2)
ax2.set_aspect(5.0)
plt.title('Differentiation of ReLU')
plt.xlabel('x')
plt.ylabel('A(x)')
plt.grid(True)
plt.plot(x, y2)

fig.savefig('diff_activation_function.png')
```
